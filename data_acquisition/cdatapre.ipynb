```python
import pandas as pd
```

```python
file1=r'data_acquisition/test.csv'
file2=r'data_acquisition/train.csv'
```

```python
df=pd.read_csv(file1)
df.head()
```

```python
nan_counts = df.isna().sum()
print("Number of NaN values per column:\n", nan_counts)
```

```python
df2=pd.read_csv(file2)
df2.head()
```

```python
nan_counts = df2.isna().sum()
print("Number of NaN values per column:\n", nan_counts)
```

```python
import pandas as pd
from scipy.stats import skew

# Load your data

# Select numeric columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate skewness for each numeric column
skewness_values = df[num_cols].apply(lambda x: skew(x.dropna()))

# Print skewness
print("Skewness values for numeric columns:")
print(skewness_values)

# Interpretation
for col, val in skewness_values.items():
    if val > 0.5:
        print(f"{col} is positively skewed.")
    elif val < -0.5:
        print(f"{col} is negatively skewed.")
    else:
        print(f"{col} is approximately symmetric (not skewed).")
```
